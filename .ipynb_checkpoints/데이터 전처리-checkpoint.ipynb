# 데이터 전처리 과정

### 적용된 작업
* **스케일링 및 필요 컬럼 재정의**
* **결측치 처리**
* **데이터 병합**
* **데이터 자료형 변환**

## 필요 모듈 IMPORT

```python
import pandas as pd
import os
```

## 타겟 데이터 파일명 가져오기

```python
files = os.listdir('data')
```

## 빈 데이터 프레임 선언

```python
data = pd.DataFrame()
type(data)
```

## 각 데이터들을 읽어들이고 병합

```python
for file in files:
    temp = pd.read_csv(f'data/{file}',encoding='ansi')
    data = pd.concat([temp,data])
data
```

## 결측치 처리

```python
data = data[data['휘발유'] != 0]
data = data[data['경유'] != 0]
data
```

## 지역컬럼을 도, 시군구 컬럼으로 분리

```python
data['도'] = data['지역'].str.split().str[0]
data['시군구'] = data['지역'].str.split().str[1]
data
```

## 기간 데이터 문자열로 변환

```python
data = data.astype({'기간':'str'})
```

## 필요컬럼 정의

```python
col_list = ['도','시군구','상호','주소','기간','상표','셀프여부','휘발유','경유','실내등유']
data = data[col_list]
data
```

## 기존 인덱스를 제거하고 새로 부여

```python
data.reset_index(drop=True,inplace=True)
data
```

```python
data.to_csv('refined_data/data.csv',encoding='ansi')
```
